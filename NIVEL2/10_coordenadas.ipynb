# Diagrama de Coordenadas Paralelas
Grfico con el que podemos comparar varias observaciones **individuales** a travs de los valores de sus variables. Es un grfico que solo consiste de mltiples Ejes Y. Cada Eje representa el rango de una variable de inters. Los valores que toman las variables de una misma observacin se conectan con lneas. De este modo podemos hallar como se encuentran relacionadas las variables.    
**Ejes Y**: Variable numrica o categrica 




### Cuando utilizar
- Quiero comparar como se relacionan diversas variables numricas de observaciones individuales. 
- Es una alternativa superior al Radar Chart, ya que no importa la escala de mis variables. 


### Consejos
- Puedo introducir una variable diferenciadora categrica que quiera explorar utilizando diferentes colores en las lneas.
- Si tengo muchos puntos puede volverse dificil de entender. Para solucionar esto puedo muestrear observaciones o jugar con las opacidades de las lneas.
- Es un grfico con usos limitados ya que facilmente se vuelve complejo de entender y catico debido a la gran cantidad de lneas.

```python
import pandas as pd
import plotly.express as px

# Algunos ambientes van a necesitar este modo para poder visualizar los graficos de Plotly
import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
df['species_id'] = df['species'].astype('category').cat.codes
df
```

```python
fig = px.parallel_coordinates(
    df, 
    color="species_id",
    dimensions=['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g'],
    color_continuous_scale=px.colors.diverging.Tealrose
)
fig.show()
```
