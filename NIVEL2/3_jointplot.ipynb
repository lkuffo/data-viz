# Joint Plot
Es un Scatter Plot o Regression Plot con histogramas (o KDE) sobre la distribucin de cada variable. Para este grfico, ambas variables deben ser numricas. 
**Eje X**: Variable numrica  
**Eje Y**: Variable numrica  



## En esta clase
### Datos
Informacin sobre especies de pinguinos
### Objetivos
1. Encontrar alguna correlacin entre las caractersticas de los pinguinos y al mismo tiempo ver la distribucin de las variables

## Referencias
- https://seaborn.pydata.org/generated/seaborn.jointplot.html

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df
```

```python
# Eliminamos filas NaNs
df = df.dropna()
df
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    kind="reg"
)
```

```python
# Jointplot Flipper lenght vs Culmen Lenght 
# https://seaborn.pydata.org/generated/seaborn.jointplot.html
sns.jointplot(
    data=df,
    x='flipper_length_mm',
    y='culmen_length_mm',
    hue='species'
)
```
