# Circle Packing (*Clase en Preparacin*)
Es hermano del TreeMap. Tiene la misma funcionalidad, las mismas ventajas y desventajas. Pero en ves de cuadrados, se utilizan crculos para agrupar las diferentes categorias. Es un grfico jerrquico sin ejes. El grfico realiza un "groupby" visual de varios niveles y muestra la variable numrica agregada resultante de cada agrupacin diferente como un crculo. El tamao de este crculo es proporcional al valor de esta variable numrica agregada. Los niveles se agrupan en crculos uno dentro del otro, siendo el primer nivel el crculo mas externo.




### Cuando utilizar
- Mi dataset tiene variables categricas que codifican grupos de una manera jerrquico y quiero conocer las diferencias entre las **magnitudes** de los valores de variables numricas agregadas dentro de estos diferentes grupos.

### Cuando no utilizar
- Si mis variables categricas jerrquicas a agrupar son temporales, por ejemplo, aos - meses - dias, no es comn hacer un treemap para estos casos. En estos casos sera preferible un grfico de lneas para ver las diferencias a travs del tiempo.
- Si quiero visualizar las diferencias de las **proporciones** entre los grupos de un mismo nivel. En este caso debemos utilizar un Sunburst.


### Consejos
- TreeMap o Circle Packing? El Circle Packing es mas atractivo visualmente. Pero no es eficiente con el espacio. Por lo que, si debemos presentar la visualizacin en algun documento formal, se prefiere el TreeMap.
- Es comn utilizar colores en el grfico para distinguir cada categora en el primer nivel. Tambin se pueden utilizar los colores para codificar la variable numrica o una segunda variable numrica relacionada a los grupos.
- Podra ser una alternativa a un Grfico de Barras con muchas categorias. Sin embargo, perdemos presicin. An mas si las magnitudes son similares entre s.
