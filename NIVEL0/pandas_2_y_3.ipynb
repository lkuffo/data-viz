```python
import pandas as pd
```

```python
# CARGA DE DATOS
df1 = pd.read_csv("./dataset_2a.csv")
df2 = pd.read_csv("./dataset_2b.csv")
df3 = pd.read_csv("./dataset_2c.csv")
```

```python
# JOINS
# Hacemos un Join para unir datos en dos datasets diferentes a partir de una columna
# El Join es una operaci√≥n que en Pandas se logra con la funci√≥n '.merge'
df_con_usuarios = df1.merge(df2, on='user', how='left')
df_con_usuarios
```

```python
# JOINS
# Volvemos a hacer un Join para obtener el nombre del pais
df_completo = df_con_usuarios.merge(df3, on='ciudad', how='left')
df_completo
```

```python
# MANEJO DE FECHAS
# Las fechas por defecto son cadenas de texto
# Pero podemos transformarlas a Fechas de Python
# Tener las fechas como Fechas de Python nos permite reformatear fechas, o ordenar el dataframe por esta columna
df_completo['fecha'] = pd.to_datetime(df_completo['fecha_publicado'], format='%d/%m/%Y')
df_completo
```

```python
# Teniendo una columna con la fecha en formato de datetime
# Puedo crear nuevas columnas que tengan esa misma fecha en nuevos formatos
# Que me permitan hacer agrupaciones interesantes
df_completo['mes'] = df_completo['fecha'].apply(lambda x: x.strftime('%m/%d'))
df_completo['dia_semana'] = df_completo['fecha'].apply(lambda x: x.strftime('%A'))
df_completo
```

```python
# Otra alternativa es, 
# Teniendo una fecha como un string ordenado, 
# Puedo parsearla y agrupar por periodos de tiempo
# como por ejemplo meses, a√±os o dias
df_completo['mes'] = df_completo['fecha_publicado'].apply(lambda x: x[3:])
df_completo
```

```python
# AGRUPACIONES MULTIPLES
# Puedo agrupar por multiples niveles en mis dataframes
agrupado = df_completo.groupby(['mes', 'ciudad']).agg({
    'id': 'count'
})
agrupado
# El nivel 0 sera mes, y el nivel 1 sera ciudad
```

```python
# PIVOTES
# A veces me es util tener cada valor diferente del 2do nivel de una agrupaci√≥n como una columna diferente
# Especialmente cuando hacemos visualizaciones
# Esto se llama pivotar una columna
# Cada VALOR diferente de la columna pasa a ser una columna del dataframe

# Si queremos pivotar la columna ciudad, necesitamos tenerla primero como una columna (y no como indice)
# .reset_index me sirve para convertir un indice, en una columna normal
desagrupado = agrupado.reset_index(1)
desagrupado
```

```python
# Ahora puedo pivotar la columna
pivote = desagrupado.pivot(columns='ciudad')
pivote
```

```python
pivote[['GUADALAJARA']]
# Porque no funciona? 
# Porque las columnas ahora tienen 2 niveles
```

```python
# Lo podemos solucionar con .droplevel, para eliminar el primer nivel de las columnas
pivote = pivote.droplevel(0, axis='columns')
pivote
```

```python
pivote[['GUADALAJARA']]
# Ahora si! 
```

```python
# Puedo llenar los na con ceros!
pivote = pivote.fillna(0)
pivote
```

```python
# Otra forma de Pivotar un dataframe, es directamente desde el indice
# Utilizando la funcion unstack()
pivote2 = agrupado.unstack(1)
pivote2
```

```python
pivote2 = pivote2.droplevel(0, axis='columns')
pivote2
# Es decir, estos dos metodos son equivalentes:
# .reset_index().pivot()
# .unstack()
```

```python
# TRANSPONER EL DATAFRAME
# Columnas se hacen filas y filas se hacen columnas
pivote2.transpose()
```
