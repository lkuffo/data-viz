# Serie de Tiempo
***(Atenction: Este no es un curso sobre anlisis de series de tiempo. Solo veremos conceptos bsicos)***   
Es un grfico de lneas cuya variable del Eje X es temporal a una granularidad muy especfica (das, horas, minutos, segundos, milisegundos). Usualmente el nmero de observaciones graficadas esta en el orden de las miles. Esto es debido a que en una serie de tiempo no nos interesan puntos individuales si no tendencias, pendientes, cambios, valles y picos a lo largo del tiempo. Es decir, analizar los puntos como un conjunto. Debido a esto, *no* se aade una clave visual como marcadores para identificar a que parte de la lnea pertenece cada punto. Pueden ser acompaados por una segunda lnea codificada en el segundo Eje Y. Tambin pueden ser acompaados por una lnea de tendencia, un intervalo de confianza de prediccin.    
**Eje X**: Variable temporal de alta granularidad   
**Eje Y**: Variable numrica     



**Ejemplos de Uso**:
- Analizar como ha cambiado el valor de la accin de una empresa a travs del tiempo

### Cuando utilizar
- Quiero analizar comportamientos a travs del tiempo:
  - Seasonality: Se denomina "Seasonality" (estacionalidad en espaol) cuando una serie de tiempo presenta caractersticas similares en periodos equivalentes de tiempo (mismo dia en diferentes aos, mismo mes en diferentes aos, mismo dia de la semana, etc). Tambien denominada patrn.
  - Tendencias: Se denomina "Tendencia" cuando observamos como a travs del tiempo el valor numrico tiende a subir o a bajar en promedio
  - Perturbaciones: Lugares en la curva donde ha ocurrido un valle o un pico que no sigue la tendencia o la estacionalidad. Son motivo de investigacin.
    - Valles: Perturbacin donde la curva disminuye de golpe pero luego de un determinado tiempo vuelve a subir.
    - Picos: Perturbacin donde la curva aumenta de golpe pero luego de un determinado tiempo vuelve a bajar

### Cuando no utilizar
- Si no tengo mediciones temporales de baja granularidad
- Si quiero analizar puntos individuales y no las tendencias, estacionalidades o perturbaciones

### Consejos
- No es necesario que el eje Y empiece en el valor del "cero" (a diferencia de los graficos de barras o de lneas comunes). Ya que no analizamos valores individuales en la curva.
- No interpolar o suavizar la curva.
- Si vamos a hacer mltiples series de tiempo en el mismo grfico: 
  - Procurar que no existan muchas lineas.
  - Procurar que las variables que cada linea representan sean diferenciables (por ejemplo, a traves de colores diferentes)
  - Procurar que ambas series de tiempo tengan rangos de valroes numricos comparables. De otro modo, una lnea quedar ilegible como una recta (ver imagen)
- Si mi numero de observaciones se encuentra en el orden de las decenas (<50), es complicado analizar una grfica de lnea como serie de tiempo.

Las series de tiempo son un tpico que merece un curso entero para ser analizadas en detalle!

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/tienda_ventas.csv')
df
```

```python
# Ventas por mes en todas las tiendas tomando en cuenta todos los productos
por_dia = df[['date', 'sales']].groupby('date').sum()
por_dia
```

```python
# Grafico de Lineas con Pandas
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
por_dia.plot.line(
    y='sales', 
    figsize=(40,10)
)
plt.title('Ventas por da')
```

```python
# Grafico de Lineas con Plotly
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
fig = px.line(
    por_dia, 
    x=por_dia.index.values,
    y="sales", 
    title='Ventas por dia'
)
fig.show()
```

```python
df2 = pd.read_csv('../data/marcas_mensajes_twitter.csv')
df2
```

```python
df2['fecha_parseada'] = pd.to_datetime(df2['created_at'], format='%a %b %d %H:%M:%S %z %Y')
df2
```

```python
df2 = df2.sort_values('fecha_parseada')
df2['periodo'] = df2['fecha_parseada'].apply(lambda x: x.strftime('%Y-%m-%d'))
df2
```

```python
df2_agrupado = df2[['periodo', 'idx']].groupby('periodo').count()
df2_agrupado = df2_agrupado.reset_index()
df2_agrupado
```

```python
# Grafico de Lineas con Plotly
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.line.html
fig = px.line(
    df2_agrupado, 
    x="periodo",
    y="idx", 
    title='Mensajes por dia'
)
fig.show()
```
