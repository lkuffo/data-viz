# Grfico de Dispersin en 3D (Scatter Plot en 3D)
Grfico que muestra los valores de **TRES** variables de un mismo registro en nuestro dataset. Aparece un nuevo Eje Z a lo profundo del grfico. En cada eje se coloca una variable. Su objetivo es detectar si existe o no alguna correlacin entre las tres variables. Los ejes son variables numricas pero tambin pueden utilizar variables categricas ordinales (aunque no es comn). El principal problema con este grfico es la percepcin de la profundidad.   
**Eje X**: Variable numrica o categrica   
**Eje Y**: Variable numrica o categrica 
**Eje Z**: Variable numrica o categrica   



**Ejemplos de Uso**:
- Como un doctor, me gustara saber si el nivel de colesterol de una persona se correlaciona con su peso

### Cuando utilizar
- Quiero comprobar si existe algun tipo de correlacin entre tres variables al mismo tiempo. 
- Estoy trabajando con datos espaciales tridimensionales

### Consejos
- Puedo introducir una 4ta variable de la siguiente manera:
  - Una variable numrica a travs de la intensidad del color de los puntos
  - Una variable categrica a travs de la forma de los puntos
- Es muy complicado percibir los efectos de la tercera variable debido a la dificultad al momento de percibir profundidad. 
- En la gran mayora de los casos un PairPlot ser siempre una mejor opcin.
- Tiene usos muy limitados. Ms utilizado en computacin cientfica




## En esta clase
### Datos
Informacin sobre especies de pinguinos
### Objetivos
1. Encontrar alguna correlacin entre las caractersticas de los pinguinos


```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/pinguinos.csv')
df = df.dropna()
df
```

```python
# Scatter plot 3D en matplotlib
# https://matplotlib.org/stable/gallery/mplot3d/scatter3d.html

fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.scatter3D(
    df['culmen_length_mm'], 
    df['culmen_depth_mm'], 
    df['flipper_length_mm'], 
    color = "green"
)

ax.set_xlabel('culmen_length_mm')
ax.set_ylabel('culmen_depth_mm')
ax.set_zlabel('flipper_length_mm')

plt.show()

# Es complicado ver la profundidad si los puntos estan dibujados en dos dimensiones
```

```python
# 3D Scatter Plot con Plotly
# https://plotly.com/python/3d-scatter-plots/
fig = px.scatter_3d(
    df, 
    x='culmen_length_mm',
    y='culmen_depth_mm', 
    z='flipper_length_mm',
    symbol='species',
    color='species'
)
fig.show()
```
