# Grfico de Linea
Grfico de valores numricos conectados por lineas de izquierda a derecha. Tiene el mismo objetivo que el grfico de barras. Nos sirve para mostrar diferencias de propiedades numricas a lo largo de diferentes grupos o puntos. Sin embargo, el eje X de izquierda a derecha debe tener una unidad de progresin ordenada de menor a mayor (ejemplo: das, aos, meses, horas; es decir, usualmente unidades de tiempo). A diferencia del grfico de barras, yo puedo tener muchas categoras o puntos en el eje X.    
**Eje X**: Variable categrica ordinal     
**Eje Y**: Variable numrica     
**Otros nombres**: Cuando la granularidad de un grfico de lnea es muy detallada, se puede denominar Serie de Tiempo.



**Ejemplos de Uso**:
- Si soy dueo de una pgina web, me gustara saber como ha cambiado el nmero de visitas entre diferentes aos
- En un banco, me interesara saber cuantas transacciones ocurren mensualmente

### Cuando utilizar
- Comparar propiedades numricas de diferentes grupos o puntos en mis datos. 
- Quiero enfatizar cambios en magnitud a lo largo de mi eje X.

### Cuando no utilizar
- Si mi variable en X no tiene una progresin. En ese caso, tendra que pensar en un grafico de barras.
- Si tengo menos de 3 puntos para graficar.
- Si quiero comparar proporciones.
- Si quiero obtener informacin sobre distribuciones.

### Consejos
- No es necesario que el eje Y empiece en el valor del "cero" (a diferencia de los graficos de barras)
- No utilices colores innecesariamente
- No interpoles o suavices la curva
- Si no hay una cantidad grande de puntos, aade una clave visual para identificar correctamente a que parte de la lnea pertenece cada punto (un triangulo, un cuadrado, un circulo, una equis, etc).
- Si vas a hacer un grfico multilinea: 
  - Procura que no existan muchas lineas.
  - Procura que las variables que cada linea representan sean diferenciables (por ejemplo, a traves de colores diferentes)

# Grfico de Area
Combina la utilidad del grfico de lneas con el grfico de barras. Me permite comparar diferencias entre magnitudes de diversos grupos a lo largo de una progresin continua. Bsicamente es un grfico de lneas con el area bajo la curva de un color slido.   
**Eje X**: Variable categrica continua   
**Eje Y**: Variable numrica   
**Otros nombres**: Area Chart
**Subtipos**: a) Stacked Area Chart (multiples lneas que acumulan sus valores numricos del eje Y) b) Overlapped Area Chart (multiples lneas que no acumulan sus valores numricos del eje Y)



## Referencias
https://chartio.com/learn/charts/area-chart-complete-guide/

```python
# Grafico de area de multiples categorias
pivote.plot.area(
    style='.-',
    stacked=True
)
```
