# Gráfico de Cajas
Muestra los cuartiles y otros estadísticos de uno o mas grupos de datos numéricos mediante cajas y líneas. La "caja" es mi delimitador de donde se encuentra el 50% de mis datos (Q1 a Q3). La línea dentro de la caja representa el valor de la mediana. Una línea adicional puede ser dibujada para el valor de la media. Los bigotes son una técnica para representar el límite de lo "normal" de mi distribución. Cualquier punto fuera de los bigotes se conoce como "datos aberrantes" (outliers).   
**Eje X**: Variable numérica   
**Eje Y**: Variable categórica    
**Otros nombres**: Grafico de Bigotes, Box Plot   






### Cuando utilizar
- Comparar grupos a través de sus estadísticos (media, mediana y cuartiles)
- Detectar datos aberrantes dentro de alguna categoría

### Cuando no utilizar
- Si yo quiero observar la **forma** exacta de mi distribución
- Si mis puntos de interés son los puntos individuales de mi dataset 

### Consejos
- Cajas verticales o horizontales? No hay preferencia. Puede depender de cuantas categorías quiero mostrar o del tamaño de mis etiquetas categóricas

### Ejemplos de Uso
- Si soy dueño de una página web, me gustaría saber cual es la diferencia del tiempo en que las personas visitan mi página desde un Celular, Tablet o PC de escritorio.
- En un banco, me interesaría saber cuanto dinero almacenan diferentes tipos de cuentas bancarias

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/titanic.csv')
df
```

```python
# Eliminar datos vacios
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
# Grafico de Cajas con Pandas (Matplotlib)
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.boxplot.html
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html#matplotlib.pyplot.boxplot
df.boxplot(
    column='Age', 
    by='Sex'
)
```

```python
# Grafico de Cajas con Seaborn
# https://seaborn.pydata.org/generated/seaborn.boxplot.html
sns.boxplot(
    df, 
    x='Sex', 
    y='Age'
).set(title='Numero de Pasajeros por Sexo')
```

# Gráfico de Violines
Los Graficos de Cajas no nos dicen nada acerca de la forma de la distribución de los datos. Solamente de los estadísticos de la distribución. Esto hace que puedan ser engañosos (ver imagen). Por esto un excelente acompañante o remplazo de los gráficos de cajas, es el gráfico de violines. Nos permite ver como estan distribuidos los datos manteniendo todas las ventajas de un diagrama de cajas. La única desventaja es que puede ser mas complejo de entender para una persona inexperta que el diagrama de cajas.


vs



**Eje X**: Variable categórica   
**Eje Y**: Variable numérica   
**Otros nombres**: Violin Plot   


### Cuando utilizar
- Quiero comparar grupos a través de sus estadísticos (media, mediana y percentiles) y sus distribuciones al mismo tiempo

### Cuando no utilizar
- Si mis puntos de interés son los puntos individuales de mi dataset
- Si **no** me interesa la forma de la distribución de mis datos, prefiero un diagrama de cajas convencional.

## Referencias
- https://datavizcatalogue.com/ES/metodos/diagrama_de_violin.html
- https://blog.bioturing.com/2018/05/16/5-reasons-you-should-use-a-violin-graph/

```python
# Violin Plot con Seaborn
# https://seaborn.pydata.org/generated/seaborn.violinplot.html
sns.violinplot(df, x='Sex', y='Age').set(title='Numero de Pasajeros por Sexo')
```

```python
sns.violinplot(
    df, 
    x='Sex', 
    y='Age', 
    inner='quartiles' # point, stick
).set(title='Numero de Pasajeros por Sexo')
```
