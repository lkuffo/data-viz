# Grfico de Barras
Un grfico simple pero efectivo. Nos sirve para mostrar diferencias de magnitudes de variables numricas a lo largo de diferentes grupos o categorias.    
**Eje X**: Variable categrica   
**Eje Y**: Variable numrica   
**Otros nombres**: Grfico de Columnas, Bar Plot   
**Variantes**: Grfico de Barras Multiples, Grfico de Barras Apiladas



### Cuando utilizar
- Comparar variables numricas de diferentes grupos en mis datos
- Quiero enfatizar cambios en magnitud

### Cuando no utilizar (recomendaciones)
- Si quiero comparar los valores de muchas categorias. Ya que el Eje X se vuelve muy denso.
- Si mis puntos de inters son los puntos individuales de mi dataset. Ya que el Eje X se volver muy denso.
- Si mi objetivo es comparar proporciones porcentuales o partes de un todo. 
- Si quiero obtener informacin sobre las distribuciones de mis variables.

### Ejemplos de Uso
- Si soy dueo de una pgina web, me gustara saber cuantas personas visitan mi pgina desde un Celular, Tablet o PC de escritorio.
- En un banco, me interesara saber el nmero de cuentas abiertas cada ao
- En un supermercado, me gustara saber el valor total de ventas de cada mes en un ao


### Consejos
- No utilices colores en las barras innecesariamente.
- Empieza el eje Y en el valor del "cero". Asi no vas a "disfrazar" cambios de magnitudes irrelevantes **(Ver primera imgen de abajo)**
- Si las variables categoricas guardan un orden (Ejemplo: aos, meses, horas), se recomienda ordenar las columnas. Del mismo modo si quieres enfatizar que es un "Top" (en cuyo caso las barras se ordenarian segun el valor en el Eje Y). 
- No hacer las barras "tridimensionales" o con formas extraas que puedan cofundir al lector acerca del valor en el eje Y **(Ver segunda imgen de abajo)**
- Barras verticales o horizontales? No hay preferencia. Puede depender de cuantas categoras quiero mostrar o del tamao de mis etiquetas categricas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leemos los datos con Pandas
df = pd.read_csv('../data/titanic.csv')
df
```

```python
# Agrupamos por Sexo y contamos cuantos registros hay en cada grupo
pasajeros = df.groupby('Sex').count()
pasajeros
```

```python
# Grafico de Barras con Pandas (matplotlib)
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.bar.html
pasajeros.plot.bar(
    y='PassengerId', 
    title='Numero de Pasajeros por Sexo',
    width=0.3
)
```

```python
# Para tener 'Sex' como una columna
pasajeros2 = pasajeros.reset_index()
pasajeros2
```

```python
# Grafico de Barras con Seaborn
# https://seaborn.pydata.org/generated/seaborn.barplot.html
sns.barplot(
    pasajeros2, 
    x='Sex', 
    y='PassengerId', 
    width=0.4
).set(title='Numero de Pasajeros por Sexo')
```

```python
# Grafico de Barras con Matplotlib sin pandas
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.bar(
    df['Sex'], 
    df['PassengerId'],
    width=0.5
)
plt.title('Numero de Pasajeros por Sexo')
```
